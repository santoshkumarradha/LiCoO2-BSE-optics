```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot(fname):
    plt.style.use('seaborn-paper')
    data=np.loadtxt(fname,comments="%")
    energy=13.6056980659*data[:,0]
    x_spinup=data[:,1]
    y_spinup=data[:,2]
    z_spinup=data[:,3]
    spin=0
    if data.shape[1]>5:
        x_spindown=data[:,4]
        y_spindown=data[:,5]
        z_spindown=data[:,6]
        spin=1

    fig,ax=plt.subplots(2,1,figsize=(13,6))
    ax[0].plot(energy,x_spinup,label="x up")
    ax[0].plot(energy,y_spinup,label="y up")
    ax[0].plot(energy,z_spinup,label="z up")
    ax[0].set_xlabel("Energy eV")
    if spin:
        ax[1].plot(energy,x_spindown,label="x down")
        ax[1].plot(energy,y_spindown,label="y down")
        ax[1].plot(energy,z_spindown,label="z down")
        ax[1].legend()
    ax[1].set_xlabel("Energy eV")
    
    ax[0].legend()
    #ax[0].set_xlim([0,6])
    #ax[1].set_xlim([0,6])
    plt.tight_layout()
    plt.show()
```

```python
plot("opt.temp")
```

```python

plt.style.use('seaborn-paper')
data=np.loadtxt("eps_rpa",comments="#")
fig,ax=plt.subplots(2,1,figsize=(13,4))

energy=data[:,0]
x_spinup=data[:,1]
y_spinup=data[:,2]


ax[0].plot(energy,x_spinup,label=r'$Re(\epsilon)$',alpha=0.4)
ax[0].plot(energy,y_spinup,label=r'$Im(\epsilon)$')
ax[0].set_xlabel("Energy eV")


data=np.loadtxt("eps_bse",comments="#")
energy=data[:,0]
x_spinup=data[:,1]
y_spinup=data[:,2]


ax[0].plot(energy,x_spinup,label=r'$Re(\epsilon) bse$',alpha=0.4)
ax[0].plot(energy,y_spinup,label=r'$Im(\epsilon) bse$')
ax[0].set_xlabel("Energy eV")

ax[0].legend()
#ax[1].legend()
#ax.set_xlim([0,6])
plt.tight_layout()
plt.show()
```
